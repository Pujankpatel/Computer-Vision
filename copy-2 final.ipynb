```python
model = YOLO("yolov8x.yaml")  # build a new model from scratch
model = YOLO("yolov8x.pt")  # load a pretrained model (recommended for training)

# Use the model
model.train(data="/kaggle/input/new-ui-component/data.yaml", epochs=80,imgsz=520,batch=10)  # train the model
 # export the model to ONNX format
metrics = model.val()
results = model("/kaggle/input/resultingui2/Screenshot 2024-03-24 192951.png")  # predict on an image
path = model.export(format="onnx")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
results1 = model.predict(source="/kaggle/input/resultingui2/Screenshot 2024-03-24 193020.png",
              save=True, conf=0.2,iou=0.5)#predicting the image from trained model and store it in result

Results = results1[0]
plot = results1[0].plot()
plot = cv2.cvtColor(plot, cv2.COLOR_BGR2RGB)
display(Image.fromarray(plot))
```
